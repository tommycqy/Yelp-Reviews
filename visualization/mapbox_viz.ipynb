```python
df = pd.read_csv(os.path.join(dir_path, 'api_data.csv')).rename(columns = {'latitude':'lat', 'longitude':'lon'})
df['name'] = df['name'].str.replace("'","")
```

```python
def get_center(df):
    length = len(df)
    sum_x = np.sum(df['lat'])
    sum_y = np.sum(df['lon'])
    
    return (sum_y/length, sum_x/length)

df_center = get_center(df)
```

```python
rest_json = df_to_geojson(df.fillna(''),
                          properties=['name', 'rating', 'price'],
                          precision=4)
```

```python
category_color_stops = [['$', 'rgb(211,47,47)'],
                        ['$$', 'rgb(81,45,168)'],
                        ['$$$','rgb(2,136,209)'],
                        ['$$$$','rgb(255,160,0)']]

viz = CircleViz(rest_json,
                access_token = access_token,
                label_property = 'name',
                color_property = 'price',
                color_function_type = 'match',
                color_stops = category_color_stops,
                center = df_center,
                zoom = 13)

viz.show()
```
